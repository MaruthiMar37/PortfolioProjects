Brief Note
Below are the names of the top 5 electronic manufacturing companies in America as suggested by http://www.electronicsandyou.com/blog/top-electronic-companies-in-usa.html (as at 15/8/2022):

1.Cisco Systems (CSCO)

2.Texas Instruments (TXN)

3.IBM (IBM)

4.Intel (INTC)

5.HP (HPE)

For easy identification, their tickers, as used in Yahoo Finance, are written in paranthesis and will be a mode of identification throughout the analysis. For each company, the balance sheet, financials sheet, and cashflow sheet will be exctracted from Yahoo Finance before analysis will be carried out on them.

The target audience for this analysis are potential investors who wish to get an indepth comparisons of the financial performances of these companies in the financial market.

Sections in this analysis include data exctraction, data cleaning, and descriptive analysis.

Please note, for the sake of this analaysis, only the year of the data retrieved is regarded (i.e the month and date of the data is discarded). This is to ensure uniformity among the data of the five companies. This may not be ideal in real life situations.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import yfinance as yf
```

```python
CSCO = yf.Ticker('CSCO')
TXN = yf.Ticker('TXN')
IBM = yf.Ticker('IBM')
INTC = yf.Ticker('INTC')
HPE = yf.Ticker('HPE')

#Assigne tickers for each company
```

# Data Extraction and wrangling of Balance Sheets

```python
CSCO_balance_sheet = CSCO.balance_sheet.transpose()   #scrap balance sheet and transpose it
CSCO_balance_sheet.index = CSCO_balance_sheet.index.year    #change the index from date type to year
CSCO_balance_sheet.index.names = ['Year']    #rename the index to 'year'
CSCO_balance_sheet = CSCO_balance_sheet.loc[::-1]    #reverse the ordering of the table from bottom to top 

TXN_balance_sheet = TXN.balance_sheet.transpose()    #scrap balance sheet and transpose it
TXN_balance_sheet.index = TXN_balance_sheet.index.year    #change the index from date type to year
TXN_balance_sheet.index.names = ['Year']     #rename the index to 'year'
TXN_balance_sheet = TXN_balance_sheet.loc[::-1]    #reverse the ordering of the table from bottom to top

IBM_balance_sheet = IBM.balance_sheet.transpose()    #scrap balance sheet and transpose it
IBM_balance_sheet.index = IBM_balance_sheet.index.year    #change the index from date type to year
IBM_balance_sheet.index.names = ['Year']     #rename the index to 'year'
IBM_balance_sheet = IBM_balance_sheet.loc[::-1]    #reverse the ordering of the table from bottom to top

INTC_balance_sheet = INTC.balance_sheet.transpose()    #scrap balance sheet and transpose it
INTC_balance_sheet.index = INTC_balance_sheet.index.year    #change the index from date type to year
INTC_balance_sheet.index.names = ['Year']     #rename the index to 'year'
INTC_balance_sheet = INTC_balance_sheet.loc[::-1]    #reverse the ordering of the table from bottom to top

HPE_balance_sheet = HPE.balance_sheet.transpose()    #scrap balance sheet and transpose it
HPE_balance_sheet.index = HPE_balance_sheet.index.year    #change the index from date type to year
HPE_balance_sheet.index.names = ['Year']     #rename the index to 'year'
HPE_balance_sheet = HPE_balance_sheet.loc[::-1]    #reverse the ordering of the table from bottom to top
```

```python
comp_list = ['CSCO', 'TXN', 'IBM', 'INTC', 'HPE']    #list of the tickers
```

```python
total_balance_sheets = pd.concat([CSCO_balance_sheet, TXN_balance_sheet, IBM_balance_sheet, INTC_balance_sheet, HPE_balance_sheet], join = 'outer', axis = 1, keys = comp_list)
total_balance_sheets
#concatenate the five sheets extracted
```

```python
total_balance_sheets.isnull().all().value_counts() #check for empty columns
```

```python
total_balance_sheets.isnull().sum() 
```

```python
total_balance_sheets.dropna(inplace=True) #dropped null values
```

# Descriptive Analysis and Possible Recommendations

# Depreciation

```python
Deprec = pd.DataFrame() #create empty data frame
```

```python
for tick in comp_list:
    Deprec[tick + ' Deprec'] = total_cashflow_sheets[tick]['Depreciation']

Deprec

#create Depreciation table
```

```python

v = Deprec.iloc[0].to_numpy()

#assign rows of the data frame (deprec)

fig, ax = plt.subplots( figsize=(8,4))
#create empty figure

ax.barh(comp_list, v, label = '2019', height = 0.6)




ax.set_ylabel("Companies", fontsize = 12, color = 'dimgray')

ax.set_xlabel("Depreciation", fontsize = 13, color = 'dimgray')

ax.set_title("Cummulative Depreciation", fontsize = 15, color = 'k')

for pos in ['right', 'top']:
    ax.spines[pos].set_visible(False)
#remove the right and top spines
    
for pos in ['left', 'bottom']:
    ax.spines[pos].set_color('lightgrey')
#change color of left and bottom spines
    
for pos in ['x', 'y']:
    ax.tick_params(pos, colors = 'dimgray')
#change color of the x and y ticks

ax.legend(loc = (1.05, 0.72))


plt.show()
```

```python
#INTC's assets has experienced a relatively high level of depreciation over the past four years,
#unlike TXN which has kept it's depreciation minimal. This means that investments in INTC will yield smaller return
#on equity for shareholdders when compared to that of other companies.
```

# Working Capital Ratio (WCR)

```python
Working_Capital_Ratio = pd.DataFrame()    #create empty data frame
```

```python
for tick in comp_list:
    Working_Capital_Ratio[tick +' WCR'] = total_balance_sheets[tick]['Total Current Assets'] - total_balance_sheets[tick]['Total Current Liabilities']

Working_Capital_Ratio

#create Working Capital Ratio table
```

```python
Working_Capital_Ratio.dropna(inplace=True)  #dropped null values
```

```python
Working_Capital_Ratio
#by removing null ,we got data of 2019,2020,2021
```

```python
fig, ax = plt.subplots(figsize=(8,4))
#create empty figure

plt.xticks([2019, 2020, 2021])
#set values of x ticks

ax.plot(Working_Capital_Ratio.index, 'CSCO WCR', data = Working_Capital_Ratio, alpha = 0.4)
ax.plot(Working_Capital_Ratio.index, 'TXN WCR', data = Working_Capital_Ratio, color = 'maroon', alpha = 0.4)
ax.plot(Working_Capital_Ratio.index, 'IBM WCR', data = Working_Capital_Ratio, alpha = 0.4)
ax.plot(Working_Capital_Ratio.index, 'INTC WCR', data = Working_Capital_Ratio, alpha = 0.4)
ax.plot(Working_Capital_Ratio.index, 'HPE WCR', data = Working_Capital_Ratio, linestyle = '-.')

ax.set_ylabel("Ratio", fontsize = 12, color = 'dimgray')

ax.set_xlabel("Years", fontsize = 13, color = 'dimgray')

ax.set_title("Yearly Working Capital Ratio", fontsize = 15, color = 'k')

for pos in ['right', 'top']:
    ax.spines[pos].set_visible(False)
#remove the right and top spines

for pos in ['left', 'bottom']:
    ax.spines[pos].set_color('lightgrey')
#change color of left and bottom spines 

for pos in ['x', 'y']:
    ax.tick_params(pos, colors = 'dimgray')
#change color of the x and y ticks

ax.legend(bbox_to_anchor = (1.3, 1))
plt.show()
```

```python
#HPE's working capital ratio has been negative over the past few years(2019 to 2021)
#and IBM to have negative working capital ratio as 2019 year year progress,,it starts to fall negative in 2020 and 2021 too
#The means that they barely have enough current assets (cash, short term investments, etc) 
#that will enable them cover the day-to-day running cost of their company.
```

# Return On Equity

```python
ROE = pd.DataFrame()    #create empty data frame
```

```python
ROE_list= ['CSCO ROE', 'TXN ROE', 'IBM ROE', 'INTC ROE', 'HPE ROE']
```

```python
for tick in comp_list:
    
    ROE[tick +' Net Worth'] = total_balance_sheets[tick]['Total Assets'] - total_balance_sheets[tick]['Total Liab']
    ROE[tick+' ROE'] = total_cashflow_sheets[tick]['Net Income']/ROE[tick+' Net Worth']
    
Return_on_Equity = ROE.loc[:, ROE_list]
Return_on_Equity

#create Return on Equity table
```

```python
Return_on_Equity.dropna(inplace=True)  #dropped null values
```

```python
Return_on_Equity
#by removing null ,we got data of 2019 alone
```

```python
v=Return_on_Equity.iloc[0]
fig, ax = plt.subplots(figsize=(8,4))


plt.xticks([2019])
#set values of x ticks
ax.barh(ROE_list, v, label = '2019', height = 0.6)

ax.set_ylabel("Return_on_Equity", fontsize = 12, color = 'dimgray')

ax.set_xlabel("Year", fontsize = 13, color = 'dimgray')

ax.set_title("Yearly Return on Equity", fontsize=14, color='k')

for pos in ['right', 'top']:
    ax.spines[pos].set_visible(False)
#remove the right and top spines

for pos in ['left', 'bottom']:
    ax.spines[pos].set_color('lightgrey')
#change color of left and bottom spines 

for pos in ['x', 'y']:
    ax.tick_params(pos, colors = 'dimgray')
#change color of the x and y ticks

ax.legend(bbox_to_anchor = (1.4, 1))
plt.show()
```

```python
#From the information on the chart, TXN has a high and stable return on equity in 2019.
#Investors are more likely to get better returns from their investment in the company more than their investments 
#in other conpanies.
```
